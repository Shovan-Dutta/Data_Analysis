```python
log_file['Dates'] = pd.to_datetime(log_file['Dates'], format="%Y-%m-%d")
log_file.plot(x="Dates", y=[" INFO"],figsize=(25, 10))
plt.show()
```

```python

log_file.plot(x="Dates", y=[" WARN","ERROR"],figsize=(25, 10))
plt.show()
```

```python
log_file.plot(x="Dates", y=["DEBUG"],figsize=(25, 10))
plt.show()
```

```python
count =[]
label = list(log_file.drop('Dates', axis=1).columns)
for i in label:
    count.append(log_file[i].sum())

# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(count, labels = label,autopct='%1.0f%%')
plt.show()
```

```python
dataplot = sb.heatmap(log_file.drop('Dates', axis=1).corr(), annot=True)
plt.show()
```
